## Exercice 1 :

On souhaite déterminer le point de la parabole $y=x^2$ le plus proche du point $M(0,2)$. 
ie: On essaye de trouver la distance minimum entre le point $M$ et la parabole $y=x^2$ 


On vas mettre ce problème sous la forme d’un problème d’optimisation sous contrainte-égalité, on a donc la fonction coût $J(x,y)$ et la contrainte $C(x,y)$ : 

$
\left\{
  \begin{array}{lr}
    \boldsymbol{J(x,y)} = \boldsymbol{(x-x_0)^2+(y-y_0)^2}  \\ 
    \boldsymbol{C(x,y)} = \boldsymbol{y-x^2} 
  \end{array}
\right.$


Avec $x_0=0$ et $y_0=2$

**1)** En premier temps on vas resoudre ce probléme en éliminant $y$ et en minimisant par rapport à $x$: 


On remplace donc $y$ avec $x^2$ et on trouve : 

$ J(x,y) = (x-x_0)^2+(x^2-y_0)^2 \\
  \implies J(x,y) = x^2+(x^2-2)^2 \\ 
  \implies J(x,y) = x^2+x^4+4-2x^2 \\
  \implies \boldsymbol{J(x,y)} =\boldsymbol{x^4-x^2+4}$

On dérive donc par rapport à $x$ :

$ \frac{\partial J(x,y)}{\partial x} = 4x^3-2x \\
 \frac{\partial J(x,y)}{\partial x} = 0 \implies 4x^3-2x=0 \\
 x (4x^2-2)=0 \implies \boldsymbol{x=0} \textrm{     ou    } \boldsymbol{(4x^2-2)=0} \\ $
 
 
 Donc on a:
 
 $ \boldsymbol{x_1=0} \\
 \boldsymbol{x_2=\sqrt{\frac{3}{2}}} \\
 \boldsymbol{x_3=-\sqrt{\frac{3}{2}}} $
 
 Grace à $y=x^2$ on trouve donc nos y : 
 
 
 $\textrm{Pour    } x_1=0                  \implies \boldsymbol{y_1=0} \\
  \textrm{Pour    } x_2=\sqrt{\frac{3}{2}} \implies \boldsymbol{y_2=\frac{3}{2}} \\ 
  \textrm{Pour    } x_3=-\sqrt{\frac{3}{2}}\implies \boldsymbol{y_3=\frac{3}{2}} \\ $
  
 

**2)** On vas resoudre ce probléme maintenant en éliminant $x$ et en minimisant par rapport à $y$: 


On remplace donc $x$ avec $\sqrt(y)$ et on trouve : 

$ J(x,y) = (\sqrt(y)-x_0)^2+(y-y_0)^2 \\
  \implies J(x,y) = y+(y-2)^2 \\ 
  \implies J(x,y) = y+y^2+4-4y \\
  \implies J(x,y) =y^2-3y+4$

On dérive donc par rapport à $y$ :

$ \frac{\partial J(x,y)}{\partial y} = 2y-3 \\
 \frac{\partial J(x,y)}{\partial y} = 0 \implies 2y-3=0 \\
 2y-3=0 \implies y=\frac{3}{2} \\ $
 
 
 Donc on a:
 
 $ y= \frac{3}{2}\\ $
 
 Grace à $y=x^2$ on trouve donc nos x : 
 
 
 $ \boldsymbol{y=\frac{3}{2} \: \implies x_1=\sqrt{\frac{3}{2}} \: \textrm{et} \: x_2=-\sqrt{\frac{3}{2}}\\} $

**3)** On vas resoudre ce probléme ensuite en utilisant les multiplicateurs de Lagrange: 

On a : 
$ L(x,y,\lambda) =J(x,y)+\lambda C(x,y)$

On trouve $x,y,\lambda$ pour $ \bigtriangledown L(x,y,\lambda) = 0$

$\boldsymbol{\bigtriangledown L(x,y,\lambda)} = 0 \implies
\left\{
  \begin{array}{lr}
    \frac{\partial L(x,y,\lambda)}{\partial x} = 2(x)-2\lambda x =0\\
    \frac{\partial L(x,y,\lambda)}{\partial y} = 2(y-2)+\lambda  =0\\
    \frac{\partial L(x,y,\lambda)}{\partial \lambda} = y-x^2 =0\\
  \end{array}
\right.$

 On trouve donc $ \boldsymbol{\lambda=1 \implies y=\frac{3}{2} \implies x_1=\sqrt{\frac{3}{2}}  \: \textrm{et} \: x_2=-\sqrt{\frac{3}{2}} }$

```python
#importation des librairies :

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
#illustration :
def J(x,y):
    return x**2+(y-2)**2
def C(x,y):
    return y-x**2 

x1d=np.linspace(-2,2,41)
y1d=np.linspace(-0.5,4,41)
x2d,y2d=np.meshgrid(x1d,y1d)

J2D=J(x2d,y2d)
```

## Exercice 5 :

Mettre en œuvre la méthode des pénalités pour déterminer le minimum de la fonction-objectif
$𝐽(x,y)=x-y$ sur le disque centré sur l’origine et de rayon 2.


On a donc la fonction coût $J(x,y)$ et la contrainte $C(x,y)$ : 

$
\left\{
  \begin{array}{lr}
    \boldsymbol{J(x,y)} = \boldsymbol{x-y}  \\ 
    \boldsymbol{C(x,y)} = \boldsymbol{x^2+y^2-4} \\ 
  \end{array}
\right.$

La méthode de pénalité consiste donc à remplacer un probléme de minimisation sous contraintes-inégalité par une suite de problémes de minimisation sans contraintes.

On a la fonction objectif avec pénalité:

$\boldsymbol{J_k(x,y)} = \boldsymbol{J(x,y)+k * max[C(x,y),0]^2} $

Et donc:


$\boldsymbol{J_k(x,y)} = \boldsymbol{x-y+k(x^2+y^2-4)^2} $

Le minimum de la fonction-objectif avec pénalité sera déterminé numériquement en utilisant la
fonction « scipy.optimize.minimize ». 

```python
#importation des librairies :
import scipy.optimize as opt
```

```python
#Definition des fonctions : 

def J(x) : 
    return x[0]-x[1]

def C(x) :
    return np.maximum(x[0]**2+x[1]**2-1,0)

def J_k(x,k) :
    return J(x)+k*C(x)**2
```

```python
x_ini=[0,0]
for k in range (10,100,10):
    J_opt= opt.minimize(J_k,x_ini,args=(k)) 
    x_min=J_opt.x
    x_ini=x_min
    print('k={} et x_min={}'.format(k,x_min))
```

On converge vers notre minimum  à chaque itérations. Donc environ : $x=-1.41456676$ et $y=1.41455455$

Normalisation du problème pour une meilleure représentation  

```python
def J(x) : 
    return x[0]-x[1]

def C(x) :
    return np.maximum(x[0]**2+(x[1]-1)**2-1,0)

def J_k(x,k) :
    return J(x)+k*C(x)**2
```

```python
#illustration :
x1d=np.linspace(-1.8,1.8,100)
y1d=np.linspace(-2,3,100)
x2d,y2d=np.meshgrid(x1d,y1d)

a=np.linspace(0,2*np.pi,180)
x_d=np.cos(a)
y_d=np.sin(a)
```

```python
plt.figure(figsize=(10,10))
fig_ini=0 
x_ini=[0,0]
for k in range (10,100,10):
    J_opt= opt.minimize(J_k,x_ini,args=(k)) 
    x_min=J_opt.x
    x_ini=x_min
    J2D=J_k([x2d,y2d],k)
    fig_ini+=1
    plt.subplot(2,2,fig_ini)
    plt.contour(x2d,y2d,J2D,60,cmap='rainbow')
    plt.plot(x_d,y_d,'k--')
    plt.plot(x_min[0],x_min[1],'rs',label='k={}'.format(k))
    plt.grid()
    plt.legend()
```
